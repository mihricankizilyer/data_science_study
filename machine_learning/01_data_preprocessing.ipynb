Data Types [2]

**Categorical**

Exp:
- smoking - not smoking
- man - woman

can not compareable

`Ordinal number`: can order, can be bigger or lower, but immeasurable

`Nominal number`: cannot order and immeasurable

**Numerirical**

`Ratio`: can be proportional, multiplied and divided. -> age

`Interval`: values that accept addition but not multiplication -> temperature
